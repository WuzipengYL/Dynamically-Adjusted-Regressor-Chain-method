```python
import pandas as pd
import numpy as np
UD_gephi = pd.read_csv('UD_gephi.csv',engine = 'python')
```

```python
grids = list(set(UD_gephi['source'].tolist()+UD_gephi['target'].tolist()))
UD_gephi_nodes = pd.DataFrame({'Id':grids})  
# 筛选出所有的城市节点，并生成dataframe
top_node = UD_gephi.sort_values(by = 'weight',ascending = False)
top_node20 = top_node['source'].drop_duplicates().iloc[:]
top_node20_df = pd.DataFrame({'Id':top_node20, 'Label':top_node20})
# 筛选出投资对数较大，且不重复的前20个城市，并生成dataframe
UD_gephi_nodes = pd.merge(UD_gephi_nodes,top_node20_df,on = 'Id',how = 'left')
# 合并，给点数据增加label字段
UD_gephi_nodes.to_csv('UD_gephi_nodes.csv',index = False)
```
